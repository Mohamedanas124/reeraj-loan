```python
df['Total_Income']=df['ApplicantIncome']+df['CoapplicantIncome']
df['Total_Loan']=df['LoanAmount']+df['Loan_Amount_Term']
```

```python
df.drop(['Loan_ID'],axis='columns',inplace=True)
```

```python
df.drop(['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term'],axis='columns',inplace=True)
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
df['Gender']=encoder.fit_transform(df['Gender'])
df['Married']=encoder.fit_transform(df['Married'])
df['Education']=encoder.fit_transform(df['Education'])
df['Self_Employed']=encoder.fit_transform(df['Self_Employed'])
df['Property_Area']=encoder.fit_transform(df['Property_Area'])
dependent={'0':0,'1':1,'2':2,'3+':3}
df['Dependents']=df['Dependents'].map(dependent)
```

```python
df['Loan_Status']=encoder.fit_transform(df['Loan_Status'])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit_transform(df)
```

```python
#correlation
corr=df.corr()
sns.heatmap(corr,annot=True)
```

```python
x=df.drop(['Loan_Status'],axis=1)
y=df['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
print(x_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(y_pred)
```
